### Linear Regression

```python
import numpy as np 
```

```python
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0.1, 0.2, 0.3, 0.5, 0.8, 2.0])

p1 = np.polyfit(x, y, 1)
p2 = np.polyfit(x, y, 2)
p3 = np.polyfit(x, y, 3)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x, y, 'ro')

xp = np.linspace(0, 5, 100)

plt.plot(xp, np.polyval(p1, xp), 'b-', label='linear regr')
plt.plot(xp, np.polyval(p2, xp), 'r--', label='quadratic, regr')
plt.plot(xp, np.polyval(p3, xp), 'g+', label='cubic regr')
plt.legend(loc= 'best')

plt.show()
```

### Non Linear Regression

```python
import numpy as np 
import matplotlib.pyplot as plt

from scipy.optimize import curve_fit

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0.1, 0.2, 0.3, 0.5, 0.8, 2.0])

p1 = np.polyfit(x, y, 1)
p2 = np.polyfit(x, y, 2)
p3 = np.polyfit(x, y, 3)

def f(x, a):
    return 0.1 * np.exp(a*x)

# print(curve_fit(f, x, y)) # print 2 values 1st one is value of a and 2nd one prints the valuse of covariance of error
# print(curve_fit(f, x, y)[0]) # this prints only value of a 
a = curve_fit(f, x, y)[0]
print(a)

plt.plot(x, y, 'ro')

xp = np.linspace(0, 5, 100)

plt.plot(xp, np.polyval(p1, xp), 'b-', label='linear regr')
plt.plot(xp, np.polyval(p2, xp), 'r--', label='quadratic, regr')
plt.plot(xp, np.polyval(p3, xp), 'g+', label='cubic regr')

plt.plot(xp, f(xp, a), 'k-', label='nonlinear regr')

plt.legend(loc= 'best')

plt.show()
```
