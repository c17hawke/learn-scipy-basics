```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
```

```python
# defining function dy/dt = -a*y 
# this is actally our diff equation that we are going to solve
def dydt(y, t ,a):
    return -a*y

a = 0.4
t = np.linspace(0, 20)
y0 = 5.0 #this is intial condition
y = odeint(dydt, y0, t, args=(a,))

plt.plot(t, y)
plt.xlabel('time')
plt.ylabel('y')
plt.show()
```
